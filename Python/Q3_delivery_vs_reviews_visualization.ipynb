```python
# -----------------------------
# Import necessary libraries
# -----------------------------
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
# -----------------------------
# File path
# -----------------------------
file_path = Path(
    r"D:\Data_Science\Projects\Portfolio_Projects\E-commerce-Sales-Delivery-and-Customer-Satisfaction\Dataset\Analysis_Outputs\Q3_delivery_vs_reviews\delivery_performance_by_review_score.csv"
)

# -----------------------------
# Load data
# -----------------------------
df = pd.read_csv(file_path)
df = df.sort_values("review_score")
```

```python
# -----------------------------
# Dark theme styling
# -----------------------------
BG = "#111111"
FG = "white"
BAR_1 = "#4cc9f0"
BAR_2 = "#f72585"

plt.rcParams["figure.facecolor"] = BG
plt.rcParams["axes.facecolor"] = BG
plt.rcParams["savefig.facecolor"] = BG
plt.rcParams["text.color"] = FG
plt.rcParams["axes.labelcolor"] = FG
plt.rcParams["xtick.color"] = FG
plt.rcParams["ytick.color"] = FG
plt.rcParams["font.size"] = 13

x = df["review_score"].astype(str)
```

```python
# -----------------------------
# Plot 1: Average delivery days by review score
# -----------------------------
fig, ax = plt.subplots(figsize=(13, 7))

bars = ax.bar(x, df["avg_delivery_days"], color=BAR_1)

ax.set_title("Average Delivery Days by Review Score", fontsize=22, pad=25, weight="bold", y=0.97)
ax.set_xlabel("")
ax.set_yticks([])
ax.grid(False)

for spine in ax.spines.values():
    spine.set_visible(False)

ax.tick_params(axis="x", labelsize=12)
ax.tick_params(axis="y", labelsize=12)

for bar, value in zip(bars, df["avg_delivery_days"]):
    ax.text(
        bar.get_x() + bar.get_width() / 2,
        value + max(df["avg_delivery_days"]) * 0.02,
        f"{value:.2f}",
        ha="center",
        va="bottom",
        fontsize=12,
        color=FG,
        weight="bold"
    )

plt.tight_layout()
plt.show()
```

```python
# -----------------------------
# Plot 2: On-time delivery rate by review score
# -----------------------------
fig, ax = plt.subplots(figsize=(12, 7))

bars = ax.bar(x, df["on_time_delivery_rate_pct"], color=BAR_2)

ax.set_title("On-Time Delivery Rate by Review Score", fontsize=22, pad=25, weight="bold", y=0.97)
ax.set_xlabel("")
ax.set_yticks([])
ax.grid(False)

for spine in ax.spines.values():
    spine.set_visible(False)

ax.tick_params(axis="x", labelsize=12)
ax.tick_params(axis="y", labelsize=12)

for bar, value in zip(bars, df["on_time_delivery_rate_pct"]):
    ax.text(
        bar.get_x() + bar.get_width() / 2,
        value + max(df["on_time_delivery_rate_pct"]) * 0.02,
        f"{value:.2f}%",
        ha="center",
        va="bottom",
        fontsize=12,
        color=FG,
        weight="bold"
    )

plt.tight_layout()
plt.show()
```
