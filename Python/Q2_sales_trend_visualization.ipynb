```python
# -----------------------------
# Import necessary libraries
# -----------------------------
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
# -----------------------------
# File path
# -----------------------------
file_path = Path(
    r"D:\Data_Science\Projects\Portfolio_Projects\E-commerce-Sales-Delivery-and-Customer-Satisfaction\Dataset\Analysis_Outputs\Q2_sales_trend\monthly_sales_and_order_trend.csv"
)

# -----------------------------
# Load data
# -----------------------------
df = pd.read_csv(file_path)
```

```python
# -----------------------------
# Dark theme styling
# -----------------------------
BG = "#111111"
FG = "white"
LINE_1 = "#4cc9f0"
LINE_2 = "#f72585"

plt.rcParams["figure.facecolor"] = BG
plt.rcParams["axes.facecolor"] = BG
plt.rcParams["savefig.facecolor"] = BG
plt.rcParams["text.color"] = FG
plt.rcParams["axes.labelcolor"] = FG
plt.rcParams["xtick.color"] = FG
plt.rcParams["ytick.color"] = FG
plt.rcParams["font.size"] = 13

x = df["year_month"]
```

```python
# -----------------------------
# Plot 1: Monthly order volume
# -----------------------------
fig, ax = plt.subplots(figsize=(13, 7))

ax.plot(x, df["total_orders"], marker="o", linewidth=2.5, color=LINE_1)

ax.set_title("Monthly Order Volume Trend", fontsize=22, pad=25, weight="bold", y=0.97)
ax.set_xlabel("")
ax.set_yticks([])
ax.grid(False)

for spine in ax.spines.values():
    spine.set_visible(False)

ax.tick_params(axis="x", rotation=45, labelsize=12)
ax.tick_params(axis="y", labelsize=12)

for i, value in enumerate(df["total_orders"]):
    ax.text(
        i,
        value + max(df["total_orders"]) * 0.015,
        f"{value:,}",
        ha="center",
        va="bottom",
        fontsize=11,
        color=FG
    )

plt.tight_layout()
plt.show()
```

```python
# -----------------------------
# Plot 2: Monthly revenue trend
# -----------------------------
def format_brl_compact(value):
    if value >= 1_000_000:
        return f"R$ {value / 1_000_000:.1f}M"
    elif value >= 1_000:
        return f"R$ {value / 1_000:.1f}K"
    else:
        return f"R$ {value:,.0f}"

fig, ax = plt.subplots(figsize=(13, 7))

ax.plot(x, df["total_revenue"], marker="o", linewidth=2.5, color=LINE_2)

ax.set_title("Monthly Revenue Trend", fontsize=22, pad=25, weight="bold", y=0.97)
ax.set_xlabel("")
ax.set_yticks([])
ax.grid(False)

for spine in ax.spines.values():
    spine.set_visible(False)

ax.tick_params(axis="x", rotation=45, labelsize=12)
ax.tick_params(axis="y", labelsize=12)

for i, value in enumerate(df["total_revenue"]):
    ax.text(
        i,
        value + max(df["total_revenue"]) * 0.015,
        format_brl_compact(value),
        ha="center",
        va="bottom",
        fontsize=10,
        color=FG
    )

plt.tight_layout()
plt.show()
```
